# Consulta de followers en Twitter y grabación en MongoDB

```python
import sys

cursor = -1

def get_followers(api, screen_name, cursor):
    """Esta función captura los seguidores de 200 en 200, haciendo la pausa cuando se cierra la ventana.
    Capturamos una selección de los campos disponibles (fields) y, si no existe el usuario, se graba en la B.D.
    """
    fields = ['id_str', 'blocking', 'description', 'created_at', 'statuses_count', 'screen_name', 'blocked_by', 
                'statuses_count', 'listed_count', 'profile_image_url', 'followers_count', 'favourites_count', 'name', 'entities', 'url']
    followers = []
    i_followers = 0
    ## cursor = -1
    params = {'screen_name': screen_name,
              'count': 200,
              'skip_status': True,
              'include_user_entities': True}
    
    while cursor != 0:   
        params['cursor'] = cursor               ## Pasar el cursor al diccionario de parámetros
        print('NUEVA PETICIÓN', i_followers)
        results = api.followers.list(**params)  ## Pasar el diccionario como si fuesen parámetros sueltos screen_name=screen_name, etc
        for r in results['users']:
            rr = {k: r[k] for k in fields}      ## Filtrado elegante de los campos haciendo un diccionario nuevo sobre la marcha
            ## followers.append(rr)
            if fwcol.count({'id_str': rr['id_str']}) == 0:    ## Prevenir la inserción duplicada
                fwcol.insert(rr)                              ## La inserción en MongoDB, mucho más cómoda que en SQL
                print(i_followers, rr['name'], ' - @' + rr['screen_name'])
            ##else:
            ##    print(i_followers, rr['name'], ' - @' + rr['screen_name'] + ' ya existe')
            sys.stdout.flush()
            i_followers += 1
        cursor = results['next_cursor']
        
    return followers
    
```

# Otro proceso: cargar los datos de RTS de mis usuarios

### Ahora ya existe la db quijote y la colección followers

```python
import pymongo
client = pymongo.MongoClient()
db = client.quijote
fwcol = db.followers
```

### Abrimos la base RETUITS en localhost (MySQL)

```python
from dbtw.sql import get_mysql
mysql = get_mysql('localhost', basedatos='retuits', verbose=False)
```

### Hay seguidores en la colección de MongoDB?

```python
fwcol.count()
```

### Consulta para la lista de seguidores (MongoDB)

```python
seguidores = [(f['id_str'], f['screen_name']) for f in fwcol.find({}, {'id_str': 1, 'screen_name': 1, '_id': 0})]
```

```python
seguidores[0]
```

### Consulta de RTS (MySQL) y actualización en MongoDB

```python
indice = 0
for s in seguidores:
    sql = "SELECT COUNT(*), MAX(CREATED_AT_RT), MIN(CREATED_AT_RT) FROM RTS WHERE ID_USR_RT = '" + s[0] + "'"
    dato = mysql.select_row(sql)
    if dato[0]:
        print(indice, s[1], dato[0], str(dato[1])[:10] if dato[1] else '-', str(dato[2])[:10] if dato[2] else '-')
    
    resultado = fwcol.update_one({'id_str': s[0]}, {"$set": {"rts.count": dato[0], "rts.fechamax": dato[1], "rts.fechamin": dato[2]}})
    indice += 1

    if indice % 100 == 0:
        print(indice)
```

#### Cada update devuelve un resultado de la operación

```python
resultado.raw_result
```

### Cuántos de mis followers han retuiteado alguna vez?

```python
fwcol.find({'rts.count': {'$gt': 0}}).count()
```

### Agrupar los followers por número de retuits

```python
puntos = []
for fa in fwcol.aggregate([
    {"$group": {"_id": "$rts.count", "count": {"$sum": 1}}}, 
    {'$sort': {'_id': -1}}
]):
    puntos.append({'usuarios': fa['count'], 'retuits': fa['_id']})
    print(fa, fwcol.find_one({'rts.count': fa['_id']}, {'_id': 0, 'screen_name': 1, 'rts.fechamax': 1}))
```

```python
puntos
```

```python
## Añadir usuarios acumulados
acumulado = 0
for f in puntos:
    acumulado += f['usuarios']
    f['acumulado'] = acumulado
```
