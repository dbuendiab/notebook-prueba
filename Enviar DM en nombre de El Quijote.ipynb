# Consulta de followers en Twitter y grabación en MongoDB

Usamos el API de Twitter y el de MongoDB. Recuperaremos los followers con el API [**followers/list**](https://dev.twitter.com/rest/reference/get/followers/list), aunque hay que montar un bucle para traerlos todos, ya que lo máximo permitido es ir de 200 en 200. Además hay que vigilar el final de la ventana, pero aquí tenemos la ventaja del parámetro **retry** que declaramos en la primera celda, y que habilita la gestión de la ventana desde la propia librería de Twitter.

````python
import sys

cursor = -1

def get_followers(api, screen_name, cursor):
    """Esta función captura los seguidores de 200 en 200, haciendo la pausa cuando se cierra la ventana.
    Capturamos una selección de los campos disponibles (fields) y, si no existe el usuario, se graba en la B.D.
    """
    fields = ['id_str', 'blocking', 'description', 'created_at', 'statuses_count', 'screen_name', 'blocked_by', 
                'statuses_count', 'listed_count', 'profile_image_url', 'followers_count', 'favourites_count', 'name', 'entities', 'url']
    followers = []
    i_followers = 0
    ## cursor = -1
    params = {'screen_name': screen_name,
              'count': 200,
              'skip_status': True,
              'include_user_entities': True}
    
    while cursor != 0:   
        params['cursor'] = cursor               ## Pasar el cursor al diccionario de parámetros
        print('NUEVA PETICIÓN', i_followers)
        results = api.followers.list(**params)  ## Pasar el diccionario como si fuesen parámetros sueltos screen_name=screen_name, etc
        for r in results['users']:
            rr = {k: r[k] for k in fields}      ## Filtrado elegante de los campos haciendo un diccionario nuevo sobre la marcha
            ## followers.append(rr)
            if fwcol.count({'id_str': rr['id_str']}) == 0:    ## Prevenir la inserción duplicada
                fwcol.insert(rr)                              ## La inserción en MongoDB, mucho más cómoda que en SQL
                print(i_followers, rr['name'], ' - @' + rr['screen_name'])
            ##else:
            ##    print(i_followers, rr['name'], ' - @' + rr['screen_name'] + ' ya existe')
            sys.stdout.flush()
            i_followers += 1
        cursor = results['next_cursor']
        
    return followers
    
````

### Consulta para la lista de seguidores (MongoDB)

Aquí hay una buena mezcla de sintaxis Python y MongoDB.

```fwcol.find({})``` es el ```SELECT * FROM fwcol```

de MongoDB. El segundo diccionario es la proyección, es decir, los campos que vamos a querer recuperar de todos los disponibles.

La parte de Python es la construcción

```seguidores = [(f1, f2) for f in <select>]```

donde generamos una lista de duplas sobre la marcha y la guardamos en la variable **seguidores**.

<br>

````python
seguidores = [(f['id_str'], f['screen_name']) for f in fwcol.find({}, {'id_str': 1, 'screen_name': 1, '_id': 0})]
````

````python
seguidores[0]
````

### Consulta de RTS (MySQL) y actualización en MongoDB

Aquí tiro de frase SQL y uso mis funciones de librería **mysql** para recuperar una sola fila de datos.

Luego actualizo mi collection de followers con la key **rt**, donde agrupo los tres datos recuperados (**count**, **fechamax**, **fechamin**).

````python
for s in seguidores:
    sql = "SELECT COUNT(*), MAX(CREATED_AT_RT), MIN(CREATED_AT_RT) FROM RTS WHERE ID_USR_RT = '" + s[0] + "'"
    dato = mysql.select_row(sql)
    if dato[0]:
        print("%-17s %5d rts, entre %s y %s" % (s[1], dato[0], str(dato[2])[:10] if dato[2] else '-', str(dato[1])[:10] if dato[1] else '-'))
    resultado = fwcol.update_one({'id_str': s[0]}, {"$set": {"rts.count": dato[0], "rts.fechamax": dato[1], "rts.fechamin": dato[2]}})
````

### Agrupar los followers por número de retuits

Para un número dado de retuits, contar cuántos followers han hecho esa cantidad de retuits. Lo suyo sería un histograma por franjas, pero con MongoDB ese tipo de agrupaciones todavía no sé cómo se materializa.

De todas maneras, ya se ve el comportamiento en plan función inversa: pocos followers con muchos RTS y muchos followers con pocos RTS.

````python
puntos = []
for fa in fwcol.aggregate([
    {"$group": {"_id": "$rts.count", "count": {"$sum": 1}}}, 
    {'$sort': {'_id': -1}}
]):
    puntos.append({'usuarios': fa['count'], 'retuits': fa['_id']})
    print(fa, fwcol.find_one({'rts.count': fa['_id']}, {'_id': 0, 'screen_name': 1, 'rts.fechamax': 1}))
````

````python
## La lista de diccionarios que he creado, con vistas a hacer una gráfica ----

puntos
````

````python
## Añadir usuarios acumulados. Convertir la lista de followers en una función escalera que vaya totalizando los anteriores.

acumulado = 0
for f in puntos:
    acumulado += f['usuarios']
    f['acumulado'] = acumulado
````

````python
## La nueva lista de puntos, con el nuevo campo acumulado

puntos
````
