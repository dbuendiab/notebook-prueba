## Cuadrícula de colores

En el espacio RGB, de 3 dimensiones, hacer una partición de 8 o 64 o 512 cubos, y contar los píxeles que pertenecen a cada cubo.

````python
MIN_COLOR = 0
MAX_COLOR = 255
BLOQUES_X_DIM = 8                                   ## Bloques por dimensión: 2, 4 o 8
BLOQUES_TOTALES = (BLOQUES_X_DIM ** 3)              ## Lo anterior al cubo: 8, 64 o 512
COLOR_STEP = int((MAX_COLOR+1) / BLOQUES_X_DIM)
COLOR_CENTER_OFFSET = int(COLOR_STEP / 2)
````

## Matrices (listas) para albergar resultados

Las tres matrices **pixeles** sirven para calcular el color promedio de todos los píxeles que hay en un bloque. Se trata de una media ponderada, cuyo valor se guarda en **pixeles_centroid**.

Las matrices **limite** eran para guardar los rangos, de cara a calcular cuándo un píxel está en un determinado bloque. El enfoque matricial es muy poco eficiente, es mejor buscar dimensión a dimensión. Posiblemente borre estas matrices si no las uso en ningún lado.

````python
import numpy as np
pixeles_num      = [[[0 for k in RANGO_DIM] for j in RANGO_DIM] for i in RANGO_DIM]
pixeles_peso     = [[[0 for k in RANGO_DIM] for j in RANGO_DIM] for i in RANGO_DIM]
pixeles_centroid = [[[0 for k in RANGO_DIM] for j in RANGO_DIM] for i in RANGO_DIM]

#limite_inf = [[[0 for k in RANGO_DIM] for j in RANGO_DIM] for i in RANGO_DIM]
#limite_sup = [[[0 for k in RANGO_DIM] for j in RANGO_DIM] for i in RANGO_DIM]
````

## Asignar cada píxel a su bloque correspondiente

Determino la posición en cada uno de los tres ejes RGB por separado.

Luego relleno las matrices que nos han de servir para calcular el centroide.

````python
## Función auxiliar para saber en qué rango de la dimensión está un determinado píxel
def get_pixel_dim_index(color):
    for i, r in enumerate(RANGO_COLOR):
        vmin, vmax = r, r+COLOR_STEP 
        if vmin <= color < vmax:
            return i
    raise Exception('Valor', r, 'fuera de límites')

## Cálculo de la ubicación del pixel. OJO!!!: ejecutar sólo una vez, si no se acumulan los valores. Para repetir, reinicializar matrices.
for pixel in data:

    ## Extración de valores desde fichero JSON
    #pr, pg, pb, n = pixel
    
    ## Extracción de valores desde fichero PNG
    n, rgb = pixel
    pr, pg, pb = rgb
    
    rgb = np.array((pr, pg, pb))

    ir = get_pixel_dim_index(pr)
    ig = get_pixel_dim_index(pg)
    ib = get_pixel_dim_index(pb)
    
    ## Los valores para la media ponderada (para cada bloque)
    pixeles_num[ir][ig][ib] += n
    pixeles_peso[ir][ig][ib] += n * rgb
````

## Calcular las coordenadas RGB ponderadas

El cociente de los dos valores anteriores da el valor RGB promedio de todos los píxeles asignados a este bloque

````python
for ir in RANGO_DIM:
    for ig in RANGO_DIM:
        for ib in RANGO_DIM:
            pixeles_centroid[ir][ig][ib] = (pixeles_peso[ir][ig][ib] / pixeles_num[ir][ig][ib]) \
                                            if pixeles_num[ir][ig][ib] \
                                            else (0,0,0)
pixeles_centroid
````

## Convertir el array de centroides en una lista (para salida gráfica)

Se trata de recorrer las matrices **pixeles_centroid** y **pixeles_num** tomando los valores RGB del centroide y la cantidad de píxeles, respectivamente, y guardándolos en la lista que usaremos para crear la imagen de la paleta.

Aprovechamos para suprimir los decimales y convertir a integer con ```int(round())```. La librería gráfica **PIL** espera que los valores RGB sean enteros.

````python
colores = []
for i in RANGO_DIM:
    for j in RANGO_DIM:
        for k in RANGO_DIM:
            
            r, g, b = pixeles_centroid[i][j][k]
            color_central = (int(round(r)), int(round(g)), int(round(b)))
                             
            numero_pixeles = pixeles_num[i][j][k]
            
            colores.append((numero_pixeles,color_central))
            
            print("%15s:" % (color_central,), "%s" % numero_pixeles, end="")
        print()
    print()
````

## Proceso de dibujo de los colores de la paleta

Básicamente se trata de que la librería **PIL** cree una nueva ```Image``` donde pondremos cada color en una subrectángulo. Lo más engorroso aquí es calcular las posiciones de cada rectángulo para que quede bonito. Además de la obsesión de parametrizar, claro.

````python
from PIL import Image

def dibujar_paleta(paleta, num_colores=None, offset=None, xsize=67, ysize=46, margin=7, bloques_linea=10, max_colores=100):
    if offset is None:
        offset = 0
    if num_colores is None:
        num_colores = len(paleta) if len(paleta) <= max_colores else max_colores
        
    ## Diseño del grid (máx 100x100)

    #xsize = 67
    #ysize = 46
    #margin = 7
    #bloques_linea = 10

    xsize_total = (xsize + margin) * bloques_linea + margin

    num_lineas, resto_bloques = divmod(num_colores, bloques_linea) 
    if resto_bloques != 0:
        num_lineas += 1

    ysize_total = (ysize + margin) * num_lineas + margin

    ## Ya teniendo el tamaño, puedo crear la imagen y poner fondo blanco
    imagen = Image.new('RGB', (xsize_total, ysize_total))
    imagen.paste((255, 255, 255), (0, 0, xsize_total, ysize_total))

    ## Cálculos de rectángulos
    i_color = offset
    i_color_final = offset + num_colores
    
    frecuencia, color = paleta[i_color]
    break_all = False

    for ypos in range(num_lineas):
        y0 = ypos * (ysize + margin) + margin
        y1 = y0 + ysize

        for xpos in range(bloques_linea):
            x0 = xpos * (xsize + margin) + margin
            x1 = x0 + xsize
            box = (x0, y0, x1, y1)

            imagen.paste(color, box)

            i_color += 1
            if i_color == i_color_final or i_color == len(paleta):
                break_all = True
                break
            frecuencia, color = paleta[i_color]

        if break_all:
            break
            
    return imagen
````
